```python
import matplotlib.pyplot as plt
from com.sbk.dspbase.filters.windowedsinc.windowed_sinc import blackman_windowed_sinc_normalized, blackman_windowed_sinc
from com.sbk.dspbase.dft import dft
from com.sbk.dspbase.polar import cart_to_polar
from IPython.display import display, Math, Latex, HTML, Markdown
import numpy as np
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Normalized Blackman windowd sinc kernel filter</font>"))
kernel_length = 32
frequency_cut_off = 0.2
t, h = blackman_windowed_sinc(kernel_length=kernel_length, cut_off=frequency_cut_off)
plt.plot(t, h)
plt.xlabel("Sample number")
plt.ylabel("Amplitude")
plt.show()
```

```python
f1 = 3
f2 = 60
t = np.linspace(start=0, stop=3, num=300)
s = np.sin(2*f1*np.pi*t) + np.sin(2*f2*np.pi*t)
plt.plot(s)
plt.show()
```

```python
re, im = dft(s)
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.show()
```

```python
filtered_s = np.convolve(s, h)
plt.plot(filtered_s)
plt.show()
```
