```python
import matplotlib.pyplot as plt
from com.sbk.dspbase.filters.windowedsinc.windowed_sinc import blackman_windowed_sinc_normalized, blackman_windowed_sinc
from com.sbk.dspbase.dft import dft
from com.sbk.dspbase.polar import cart_to_polar
from IPython.display import display, Math, Latex, HTML, Markdown
import numpy as np
from numpy import cos, sin, pi, absolute, arange
from scipy.signal import kaiserord, lfilter, firwin, freqz
from pylab import figure, clf, plot, xlabel, ylabel, xlim, ylim, title, grid, axes, show
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Normalized Blackman windowd sinc kernel filter</font>"))
kernel_length = 64
frequency_cut_off = 0.6
x = blackman_windowed_sinc_normalized(kernel_length=kernel_length, cut_off=frequency_cut_off)
plt.plot(x)
plt.xlabel("Sample number")
plt.ylabel("Amplitude")
plt.show()
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Frequency response of truncated sinc kernel filter</font>"))
re, im = dft(x)
ro, phi = cart_to_polar(re,im)
plt.plot(ro[0:40])
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
