```python
import matplotlib.pyplot as plt
from com.sbk.dspbase.filters.windowedsinc.windowed_sinc import sinc,blackman_windowed_sinc
from com.sbk.dspbase.dft import dft
from com.sbk.dspbase.polar import cart_to_polar
from IPython.display import display, Math, Latex, HTML, Markdown
import numpy as np
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Truncated sinc kernel filter</font>"))
kernel_length = 32
frequency_cut_off = 0.2
t, x = sinc(kernel_length=kernel_length, cut_off=frequency_cut_off)
plt.plot(t, x)
plt.xlabel("Sample number")
plt.ylabel("Amplitude")
plt.show()

```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Frequency response of truncated sinc kernel filter</font>"))
re, im = dft(x)
ro, phi = cart_to_polar(re,im)
plt.plot(ro[0:40])
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Blackman windowd sinc kernel filter</font>"))
kernel_length = 32
frequency_cut_off = 0.2
t, x = blackman_windowed_sinc(kernel_length=kernel_length, cut_off=frequency_cut_off)
plt.plot(t, x)
plt.xlabel("Sample number")
plt.ylabel("Amplitude")
plt.show()
```

```python
display(HTML("<span style='font-family:tahoma;font-size:18;color:green'>Frequency response of Blackman windowd sinc kernel filter</font>"))
re, im = dft(x)
ro, phi = cart_to_polar(re,im)
plt.plot(ro[0:40])
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
