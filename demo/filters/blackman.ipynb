```python
import matplotlib.pyplot as plt
import numpy as np
from com.sbk.dspbase.filters.blackman import blackman_kernel
import scipy.signal as signal
```

```python
f = blackman_kernel(filter_length=64)
plt.plot(f)
plt.title("Blackman window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
plt.show()
```

```python
F = np.fft.fft(f, 512) / (len(f)/2.0)
freq = np.linspace(-0.5, 0.5, len(F))
response = 20 * np.log10(np.abs(np.fft.fftshift(abs(F) / abs(F).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the Blackman window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.show()
```

```python
a = 10
t = np.linspace(start=0, stop=40, num=1000)
x = a*np.sin(t)
noise = np.random.normal(0, 1, t.shape)
s = x + noise
plt.plot(s)
plt.show()
```

```python
f_s = np.convolve(s, f)
plt.plot(f_s)
plt.show()
```
