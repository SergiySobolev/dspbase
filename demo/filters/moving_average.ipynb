```python
import matplotlib.pyplot as plt
import numpy as np
from com.sbk.dspbase.filters.moving_average import moving_average_kernel
import scipy.signal as signal
from com.sbk.dspbase.polar import cart_to_polar
```

```python
a = 10
t = np.linspace(start=0, stop=10, num=1000)
x = a*np.sin(t)
noise = np.random.normal(0, 1, t.shape)
s = x + noise
plt.plot(s)
plt.show()
```

```python


f5 = moving_average_kernel()
f9 = moving_average_kernel(filter_length=9)
f21 = moving_average_kernel(filter_length=21)

w5, h5 = signal.freqz(f5)
ro5, phi5 = cart_to_polar(h5.real, h5.imag)
plt.plot(ro5, label='Moving average 5')

w9, h9 = signal.freqz(f9)
ro9, phi9 = cart_to_polar(h9.real, h9.imag)
plt.plot(ro9, label='Moving average 9')

w21, h21 = signal.freqz(f21)
ro21, phi21 = cart_to_polar(h21.real, h21.imag)
plt.plot(ro21, label='Moving average 21')
plt.title("Frequency response of moving average filter kernels of different size")
plt.legend()
plt.show()

```

```python
r5 = signal.convolve(s, f5)
plt.plot(r5)
plt.show()
```

```python
r9 = signal.convolve(s, f9)
plt.plot(r9)
plt.show()
```

```python
r21 = signal.convolve(s, f21)
plt.plot(r21)
plt.show()
```
