```python
SN = 100000
st = np.arange(start=0, stop=SN, step=0.1)
N = 64
F = 5

t = st[0:N-1]
x = np.sin(F*t)
plt.plot(x)
plt.show()
```

```python

re_x, im_x = dft(x)
plt.plot(re_x)
plt.show()
```

```python
from com.sbk.dspbase.dft import i_dft

x_restored = i_dft(re_x, im_x)
plt.plot(x_restored)
plt.show()

```
