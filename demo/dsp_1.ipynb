```python
import numpy as np
import matplotlib.pyplot as plt
from com.sbk.dspbase.dft import dft, i_dft
from math import pi
```

```python
SN = 100000
st = np.arange(start=0, stop=SN, step=0.01)
N = 128
F = 10

t = st[0:N-1]
x = np.sin(2*pi*F*t)
plt.plot(x)
plt.show()
```

```python
re_x, im_x = dft(x)
plt.stem(re_x)
plt.show()
```

```python
x_restored = i_dft(re_x, im_x)
plt.plot(x_restored)
plt.show()
```

```python
plt.stem(abs(np.fft.fft(x))[0:N//2])
plt.show()
```
