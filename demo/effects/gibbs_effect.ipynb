```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
x = np.zeros(768)
x[256:512] = 1
re, im = dft(x)

i = widgets.IntSlider(min=0, max=re.size, step=1, value=re.size / 2)

fig, ax = plt.subplots(1, figsize=(10, 4))
plt.suptitle("Restored signal with Gibbs effect")


def update_restored_signal(bound):
    ax.clear()
    re_ = np.copy(re)
    re_[bound:] = 0
    x_restored = i_dft(re_, im)
    units = 'Higher DFT non-zero bound={}'
    ax.plot(x_restored, label=units.format(bound))
    ax.legend(loc=3)
    plt.show()



widgets.interactive(update_restored_signal, bound=i)
```
