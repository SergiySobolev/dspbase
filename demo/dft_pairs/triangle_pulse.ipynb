```python
import numpy as np
import matplotlib.pyplot as plt
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
from com.sbk.dspbase.triangle import triangle_pulse
```

```python
b = 128
N = 256
S = 128
PH = 100
t = np.linspace(0, b-1, N-1)
x = triangle_pulse(shift=S, peak_height=PH)
plt.plot(x)
plt.show()
```

```python
re, im = dft(x)
plt.plot(re[0:100])
plt.show()
```

```python
ro, phi = cart_to_polar(re, im)
plt.plot(ro[0:25])
plt.show()
```

```python

x_restored = i_dft(re, im)
plt.plot(x_restored)
plt.show()
```
