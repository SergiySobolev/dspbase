```python
import numpy as np
import matplotlib.pyplot as plt
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
```

```python
b = 128
N = 512
t = np.linspace(0, b-1, N-1)
x = np.sinc(t-b/2)


plt.plot(t, x)
plt.show()
```

```python
re, im = dft(x)
plt.plot(re)
plt.show()
```

```python
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.show()
```

```python
plt.plot(phi)
plt.show()
```

```python

X = np.fft.fft(x)
plt.plot(X.real[0:N//2])
plt.show()
```

```python
x_restored = i_dft(re, im)
plt.plot(x_restored)
plt.show()
```
