```python
import matplotlib.pyplot as plt
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
from com.sbk.dspbase.pulses.gaussian import gaussian_pulse
```

```python
t1, t2 = gaussian_pulse(deviation=1)
plt.plot(t1, t2)
plt.show()
```

```python
re, im = dft(t2)
ro, phi = cart_to_polar(re, im)
plt.plot(ro[0:50])
plt.show()
```

```python
t2_restored = i_dft(re, im)
plt.plot(t1, t2_restored[0:t1.size])
plt.show()
```
