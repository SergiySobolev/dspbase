```python
import matplotlib.pyplot as plt
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
from com.sbk.dspbase.pulses.gaussian import gaussian_burst
import scipy.signal as signal
```

```python

t1, t2 = gaussian_burst(mean=5)
plt.plot(t1, t2)
plt.show()
```

```python
re, im = dft(t2)
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.show()
```

```python
x_restored = i_dft(re, im)
plt.plot(x_restored)
plt.show()
```

```python
w,h = signal.freqz(t2)
ro, phi = cart_to_polar(h.real, h.imag)
plt.plot(ro)
plt.show()
```
