```python
import matplotlib.pyplot as plt
import numpy as np
from com.sbk.dspbase.dft import dft, i_dft
from com.sbk.dspbase.polar import cart_to_polar
from com.sbk.dspbase.pulses.gaussian import gaussian_burst
```

```python

t1, t2 = gaussian_burst(mean=5)
plt.plot(t1, t2)
plt.show()
```

```python
re, im = dft(t2)
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.xlabel('Freq')
plt.ylabel('Value)')
plt.title('Polar DFT real part')
plt.grid(True)
plt.show()
```

```python
def dilute_by_zeros(x, n=1):
    dx = np.zeros(x.size*(n + 1))
    for i in range(x.size):
        dx[i*(n+1)] = x[i]
    return dx    


```

```python

d_x = dilute_by_zeros(x=t2, n=1)
plt.plot(d_x)
plt.xlabel('Time')
plt.ylabel('Signal value')
plt.title('Diluted by 1 zero')
plt.grid(True)
plt.show()
re, im = dft(d_x)
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.xlabel('Freq')
plt.ylabel('Value)')
plt.title('Polar DFT diluted by 1 zero real part')
plt.grid(True)
plt.show()
```

```python
d_x = dilute_by_zeros(x=t2, n=2)
plt.plot(d_x)
plt.xlabel('Time')
plt.ylabel('Signal value')
plt.title('Diluted by 2 zeros')
plt.grid(True)
plt.show()
re, im = dft(d_x)
ro, phi = cart_to_polar(re, im)
plt.plot(ro)
plt.xlabel('Freq')
plt.ylabel('Value)')
plt.title('Polar DFT diluted by 2 zeros real part')
plt.grid(True)
plt.show()
```
